```python
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
from scipy.io import wavfile
from IPython.display import Audio
import librosa
import librosa.display
from scipy.io import wavio
```

```python
data = 'starbucks_a.wav'
train='starbucks_b.wav'

data1, sr1 = librosa.load(data)
n_train, sr_train = librosa.load(train)
#n_test, sr_test = librosa.load(test)

print(data1)
```

```python
Audio(data=data1, rate=sr1)
```

```python
Audio(data=n_train, rate=sr_train)
```

```python
data1_t=data1.T
origin_dot_transpose=data1_t@data1
ot=1/(origin_dot_transpose)
theta_opt=ot*data1_t@n_train

print(theta_opt)

print(np.linalg.norm(n_train))
print(np.linalg.norm(theta_opt*data1*))
```

```python
noise_cancelled_train = 0.2*data1-n_train
plt.figure(figsize=(12,4), dpi=100)
plt.plot(np.arange(len(n_train))/sr_train, n_train, label="noisy train data")
plt.plot(np.arange(len(n_train))/sr_train, noise_cancelled_train, label="noise-cancelled train data")
plt.legend()
plt.show()
Audio(noise_cancelled_train, rate=Fs, autoplay=True)
```
